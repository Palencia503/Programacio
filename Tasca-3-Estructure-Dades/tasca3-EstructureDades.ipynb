* Comprehensions, list comprehensions, set comprehensions, dict

**list comprehension** nos permiten crear listas de elementos en una sola línea de código. 
Por ejemplo, podemos crear una lista con los cuadrados de los primeros 5 números de la siguiente forma:

```python
#List Comprehensions:
cuadrados = [i**2 for i in range(5)]
print('Ejemplo: List Comprehensions')
print(cuadrados)
```

Las **Set Comprehensions** son muy similares a las listas que hemos visto con anterioridad. La única diferencia es que debemos cambiar el () por {}. Como resulta evidente, dado que los set no permiten duplicados, si intentamos añadir un elemento que ya existe en el set, simplemente no se añadirá.

```python
#Set Comprehensions:
frase = "Hola, Buenos dias"
mi_set = {i for i in frase if i == ","}
print(mi_set)
```

**Dict Comprehensions** Creemos un diccionario donde las claves sean números del 1 al 5, y los valores sean sus cuadrados:

```python
#dict comprehensions
numeros = [1, 2, 3, 4, 5]
cuadrados = {n: n**2 for n in numeros}
print(cuadrados)
```

* enumerate(), map(), filter() i reduce()

**Enumerate()** es una función de Python que añade un contador a un iterable, como una lista o una tupla, para poder acceder tanto al índice como al valor de cada elemento en cada iteración

```python
#Enumerate ejemplo
a = ["Uvas", "Fresas", "Manzanas"]

for i, name in enumerate(a):
    print(f"Index {i}: {name}")
```

**Map()** aplica una función a cada elemento de un iterable (como una lista) y devuelve un iterador con los resultados. Es una forma concisa de transformar elementos sin necesidad de un bucle

```python
#Map ejemplo
def doblar(numero):
  return numero * 2
numeros = [2, 5, 10, 23, 50]
resultado = map(doblar, numeros)
print(list(resultado))

```

**Filter()** crea un nuevo iterador que selecciona elementos de un iterable (como una lista o tupla) basándose en una función 

```python
#filter ejemplo
def par(num):
    return num % 2 == 0

numeros = [1, 2, 3, 4, 5, 6]
result = filter(par, numeros)
print(list(result))
```

**Reduce()** es una función del módulo functools que aplica de forma acumulativa una función a los elementos de un iterable, ya sean listas, tuplas o cualquier otro elemento, reduciéndolos a un solo valor

```python
#reduce ejemplo
from functools import reduce

numeros = [1, 2, 3, 4, 5]
resultado = reduce(lambda a, b: a + b, numeros)
print(resultado)
```

* Rang (estadística): diferència entre el màxim i el mínim d’una llista numèrica.

El **rango** estadistico de una lista numerica es la diferencia entre el valor maximo y el minimo, que se calcula facilmente usando las funciones incorporadas max() y min(). Simplemente llama a max() para obtener el numero mas alto y a min() para obtener el mas bajo, luego resta el minimo del maximo para obtener el rango.

```python
numeros = [10, 25, 5, 42, 18, 33]

#el valor maximo
maximo = max(numeros)

#el valor minimo
minimo = min(numeros)

#la diferencia (rango)
rango = maximo - minimo

#resultados
print(f"La lista es: {numeros}")
print(f"El valor maximo es: {maximo}")
print(f"El valor minimo es: {minimo}")
print(f"El rango de la lista es: {rango}")
```

* Quants números hi ha?
* Quantes vegades apareix el número 3.
* Quantes vegades apareixen els nombres 3 i 4?

```python
import random

#genera lista aleatoria con numeros repetidos
lista = []
for i in range(20):
    lista.append(random.randint(1, 10))

print("Lista original:", lista)

print("Cuantos numeros hay?:", len(lista))
print("Cuantas veces aparece el numero 3?:", lista.count(3))
print("Cuantas veces aparece el 3 i 4?:", lista.count(3) + lista.count(4))
print("Numero mas grande:", max(lista))
print("Los 3 numeros mas pequeños:", sorted(lista)[:3])
print("Rango de la lista:", max(lista) - min(lista))

#ordenar y elimina duplicados
llista_final = sorted(set(lista))
print("Lista final sin repeticiones:", llista_final)
```

Crea un diccionari de la següent forma i respon a les preguntes: compra = {
"Pomes" : {"Qty": 5, "€": 0.42}, "Peres" : {"Qty": 3, "€":
0.66} }

* Afegeix alguna fruita més
* Quant han costat les peres en total?
* Quantes fruites hem comprat en total?
* Quina és la fruita més cara?

Afegeix una funció que calculi el cost total de la compra i el preu mitjà per fruita.”
“Mostra la fruita més cara utilitzant una cerca automàtica amb max() i lambda.

**Suggeriments (mètodes Python):**
dict.get(), items(), keys(), values(), recorreguts amb for,
sum(), max(seq, key=lambda ...).

```python
compra = {
    "Pomes": {"Qty": 5, "€": 0.42},
    "Peres": {"Qty": 3, "€": 0.66},
    "Platanos": {"Qty": 6, "€": 0.35}
}

cost_peres = compra["Peres"]["Qty"] * compra["Peres"]["€"]
print("Coste total de las Peres:", round(cost_peres, 2), "€")

total_frutas = 0
for datos in compra.values():
    total_frutas += datos["Qty"]
print("Total de frutas compradas:", total_frutas)

fruta_mas_cara = None
precio_max = 0
for fruta, datos in compra.items():
    if datos["€"] > precio_max:
        precio_max = datos["€"]
        fruta_mas_cara = fruta
print("Fruta mas cara:", fruta_mas_cara, "con precio de", precio_max, "€")

def resumen_compra(compra):
    coste_total = 0
    total_unidades = 0
    for datos in compra.values():
        coste_total += datos["Qty"] * datos["€"]
        total_unidades += datos["Qty"]
    precio_medio = coste_total / total_unidades
    return coste_total, precio_medio

coste_total, precio_medio = resumen_compra(compra)
print("Coste total de la compra:", round(coste_total, 2), "€")
print("Precio medio por fruta:", round(precio_medio, 2), "€")
```

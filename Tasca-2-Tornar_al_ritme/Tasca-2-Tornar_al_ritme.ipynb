12. Nombre primer
* Funció que comprovi si un número és primer.

```python
def es_primer(n):
    if n <= 1:
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = int(input("Introduce un numero: "))

if es_primer(num) == True:
    print(f"{num} es primer.")
else:
    print(f"{num} no es primer.")

```

13. Calculadora modular
* Funcions separades per sumar, restar, multiplicar i dividir.

```python
def multiplicacion(num1, num2):
    m = num1 * num2
    return m
 
def division(num1, num2):
    d = num1 / num2
    return d

def resta(num1, num2):
    r = num1 - num2
    return r

def suma(num1, num2):
    s = num1 + num2
    return s

num1 = int(input('Numero 1: '))
num2 = int(input('Numero 2: '))
operacion = str(input('Tipo de calculo. ej. *, /, -, +: '))

if operacion == '*':
    resultado_m = multiplicacion(num1, num2)
    print(resultado_m)

elif operacion == '/':
    resultado_d = division(num1, num2)
    print(resultado_d)

elif operacion == '-':
    resultado_r = resta(num1, num2)
    print(resultado_r)

elif operacion == '+':
    resultado_s = suma(num1, num2)
    print(resultado_s)
    
```

14. Palindroms

```python
def es_palindrom(texto):
    texto = texto.lower()
    texto = texto.replace(" ", "")
    return texto == texto[::-1]

frase = input("Escribe una palabra o frase: ")

if es_palindrom(frase):
    print("Es un palindromo.")
else:
    print("No es un palindromo.")

```

16. Comptador de vocals
* Donar un text i retornar un diccionari amb la freqüència de cada
vocal.

```python
def comptar_vocals(text):
    vocals = "aeiou"
    comptador = {}

    for lletra in text.lower():
        if lletra in vocals:
            if lletra in comptador:
                comptador[lletra] += 1
            else:
                comptador[lletra] = 1

    return comptador

text = "Hoy hace sol"
resultado = comptar_vocals(text)
print(resultado)
```

18. Classe Compte Bancari
* Atributs: titular, saldo.
* Mètodes: ingressar, retirar, mostrar saldo.

```python
class Compte_Bancari:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    def ingresar(self, cantidad):
        self.saldo += cantidad
        print(f"Has ingresado {cantidad}€. Saldo actual: {self.saldo}€")

    def retirar(self, cantidad):
        if cantidad <= self.saldo:
            self.saldo -= cantidad
            print(f"Has retirado {cantidad}€. Saldo actual: {self.saldo}€")
        else:
            print("No tienes suficiente dinero en la cuenta.")

    def mostrar(self):
        print(f"Titular: {self.titular} | Saldo: {self.saldo}€")


titular_compte = Compte_Bancari("Jose", 20)

while True:
    operador = input("i = ingresar / r = retirar / m = mostrar / x = sortir: ").lower()
    if operador == "i":
        cantidad_ingresar = float(input("Cantidad a ingresar: "))
        titular_compte.ingresar(cantidad_ingresar)

    elif operador == "r":
        cantidad_retirar = float(input("Cantidad a retirar: "))
        titular_compte.retirar(cantidad_retirar)

    elif operador == "m":
        titular_compte.mostrar()

    elif operador == "x":
        print("saliendo...")
        break

    else:
        print("Opcion no valida.")

```

19. Classe Cotxe
* Atributs: marca, model, velocitat.
* Mètodes: accelerar, frenar, mostrar estat.

```python
class Cotxe:
    def __init__(self, marca, model, velocitat):
        self.marca = marca
        self.model = model
        self.velocitat = velocitat

    def accelerar(self, cantidad):
        self.velocitat += cantidad
        print(f"Has acelerado {cantidad} km/h. Velocidad actual: {self.velocitat} km/h")

    def frenar(self, cantidad):
        if cantidad > self.velocitat:
            self.velocitat = 0
            print("Has frenado completamente. Velocidad actual: 0 km/h")
        else:
            self.velocitat -= cantidad
            print(f"Has frenado {cantidad} km/h. Velocidad actual: {self.velocitat} km/h")

    def mostrar_estado(self):
        print(f"Marca: {self.marca} | Modelo: {self.model} | Velocidad: {self.velocitat} km/h")

un_cotxe = Cotxe("Seat", "Ibiza", 200)

print("a = acelerar / f = frenar / e = estado / x = salir")
while True:
    operador = input(": ")
    if operador.lower() == "a":
        cantidad = float(input("¿Cuanto quieres acelerar?: "))
        un_cotxe.accelerar(cantidad)

    elif operador.lower() == "f":
        cantidad = float(input("¿Cuanto quieres frenar?: "))
        un_cotxe.frenar(cantidad)

    elif operador.lower() == "e":
        un_cotxe.mostrar_estado()

    elif operador == "x":
        print("Saliendo...")
        break
    else:
        print("Opcion no valida.")
```

20. Classe Biblioteca
* Classe Llibre (titol, autor, disponible).
* Classe Biblioteca amb mètodes per afegir llibres, prestar i retornar.

```python
class Llibre:
    def __init__(self, titol, autor):
        self.titol = titol
        self.autor = autor
        self.disponible = True

class Biblioteca:
    def __init__(self):
        self.llibres = []

    def afegir(self, titol, autor):
        llibre = Llibre(titol, autor)
        self.llibres.append(llibre)
        print(f"Agregar libro: {titol}")

    def prestar(self, titol):
        for llibre in self.llibres:
            if llibre.titol == titol and llibre.disponible:
                llibre.disponible = False
                print(f"Libro prestado: {titol}")
                return
        print("No encontrado.")

    def retornar(self, titol):
        for llibre in self.llibres:
            if llibre.titol.lower() == titol.lower() and not llibre.disponible:
                llibre.disponible = True
                print(f"Libro devuelto: {titol}")
                return
        print("No encontrado.")

    def mostrar(self):
        print("Llibres a la biblioteca:")
        for llibre in self.llibres:
            if llibre.disponible:
                estat = "Disponible"
            else:
                estat = "Prestado"
            print(f"- {llibre.titol} - {llibre.autor} ({estat})")

biblio = Biblioteca()

print("a = afegir / p = prestar / r = retornar / m = mostrar / x = sortir")
while True:
    accio = input(": ").lower()
    if accio == "a":
        titol = input("Titol: ")
        autor = input("Autor: ")
        biblio.afegir(titol, autor)

    elif accio == "p":
        titol = input("Titol a prestar: ")
        biblio.prestar(titol)

    elif accio == "r":
        titol = input("Titol a retornar: ")
        biblio.retornar(titol)

    elif accio == "m":
        biblio.mostrar()

    elif accio == "x":
        print("Saliendo...")
        break
    else:
        print("Opcion no valida.")
```

21. Mini-joc POO
* Classe Jugador amb punts de vida.
* Classe Monstre amb dany d’atac.
* Simular un petit combat amb torns.

```python
import random

class Jugador:
    def __init__(self):
        self.nom = 'Jose'
        self.vida = 100
        self.dany = random.randint(15, 25)
    
    def ha_acertado(self):
        return random.randint(1, 3) == 1
    
    def atacar(self, objetivo):
        objetivo.vida -= self.dany
        print(f"{objetivo.nom} recibe {self.dany} de daño. Vida restante: {objetivo.vida}")

class Monstre:
    def __init__(self):
        self.nom = 'Titan'
        self.vida = 100
        self.dany = random.randint(15, 25)
    
    def ha_acertado(self):
        return random.randint(1, 3) == 2
    
    def atacar(self, objetivo):
        objetivo.vida -= self.dany
        print(f"{objetivo.nom} recibe {self.dany} de daño. Vida restante: {objetivo.vida}")

jugador = Jugador()
monstre = Monstre()

print("¡Comienza el combate!")
input("Presiona ENTER para comenzar")

while jugador.vida > 0 and monstre.vida > 0:
    pegar_m = input('¿Quieres pegarle al monstruo(s/n)? ')
    if pegar_m.lower() == 's':
        if jugador.ha_acertado():
            print('                         o')
            print(f'{jugador.nom} LE ESTAS PEGANDO A {monstre.nom}')
            print('-----------------------------------------------')
            jugador.atacar(monstre)
        else:
            print('No has acertado el ataque.')

        if monstre.vida <= 0:
            print('¡Has vencido al monstruo!')
            break

        if monstre.ha_acertado():
            print('                         o')
            print(f'{monstre.nom} ESTA PEGANDO A {jugador.nom}')
            print('-------------------------------------------')
            monstre.atacar(jugador)
        else:
            print('El monstruo ha fallado el ataque.')

        if jugador.vida <= 0:
            print('¡Has sido derrotado por el monstruo!')
            break
    else:
        print('No has atacado este turno.')
```
